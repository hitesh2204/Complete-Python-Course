## What is Pandas group by Function?

## Pandas group by function is used to spilt the data into groups based on some criteria.

## Syntax=DataFrame.groupby()
Group by Involed some operation.
1) Splitting the object.
2) Applying the function.
3) Combining the result.

```python
import pandas as pd
wea_df=pd.read_csv("E://Datasets//weather.csv")
wea_df
```

```python
grp1=wea_df.groupby('City')
grp1
```

```python
for City ,city_df in grp1:
    print(City)
    print(city_df)
```

```python
grp1.get_group('Pune')
```

```python
grp1.max()
```

```python
grp1.sum()
```

```python
grp1.min()
```

## What is Pandas Merge function?


## Pandas merge connects columns or indexes in Dataframe based on one o more key.

```python
import pandas as pd
df1=pd.DataFrame({'ID':[1,2,3,4,5],
                 'Age':[29,30,27,28,31]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['Hitesh','Vivek','Kunal','Rajuu']})
df2
```

```python
merge=pd.merge(df1,df2,on='ID')
merge
```

```python
merge=pd.merge(df2,df1,on='ID')
merge
```

```python
merge=pd.merge(df2,df1,on='ID',how='left')
merge
```

```python
merge=pd.merge(df2,df1,on='ID',how='right')
merge
```

```python
merge=pd.merge(df2,df1,on='ID',how='right')
merge
```

```python
merge=pd.merge(df2,df1,on='ID',how='outer',indicator=True)
merge
```

```python
df1
```

```python
df2=pd.DataFrame({'ID':[6,7,8,9],
                 'Name':['Hitesh','Vivek','Kunal','Rajuu']})
df2
```

```python
merge_df=pd.merge(df1,df2,left_index=True,right_index=True)
merge_df
```
